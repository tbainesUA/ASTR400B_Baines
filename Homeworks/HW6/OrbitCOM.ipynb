```python
"""Import Modules"""
import numpy as np
import astropy.units as u

"""Import other programs"""
from Readfile import Read
from CenterOfMass import CenterOfMass
```

```python
def OrbitCOM(galaxy, start, end, n):
    # Function that takes in inputs: Galaxy of which files to examine, the snapshot start and end(first and last), 
    # and n intervals
    # use n = 5 for HW6
    
    #Define file that will store the orbit
    fileout = "Orbit"+galaxy+".txt"
    
    #Array that will store, position and velocity coordinates along with snapshot
    Row = int(end/n)+1
    Column = 7
    Orbit = np.zeros((Row, Column))
    #print (Orbit.shape)
    
    #Define values for Center of mass
    delta = 1
    VolDec = 5
    
    
    # For loop to make Orbit file of all snap shoots of galaxy:
    for i in np.arange(start, end+n, n):
        print (i)
        # Locate file at snap of ith index
        ilbl = '000' + str(i)
        
        # Remove all but last 3 digitis
        ilbl = ilbl[-3:]
        filename = '%s_' % (galaxy) + ilbl
        
        # Define CenterOfMass class (disk particles: ptype = 2)
        Galaxy_COM = CenterOfMass(filename, 2)
        
        # Galaxy COM position calculation of ith snap shot
        G_COMx, G_COMy, G_COMz = Galaxy_COM.COM_P(VolDec, delta)
        
        # Galaxy COM velocity calculation of ith snap shot
        G_COMvx, G_COMvy, G_COMvz = Galaxy_COM.COM_V(G_COMx, G_COMy, G_COMz)
        
        # Store Quanitites w/ divided out units
        G_Time = Galaxy_COM.time/1000/u.Myr
        
        #print (G_Time)
        #print (G_COMx)

        # Identify row to apply new files data values
        ith_row = int(i/n)
        
        # Add file data to ith row of Orbit
        Orbit[ith_row, 0] = float(G_Time)
        Orbit[ith_row, 1] = float(G_COMx)
        Orbit[ith_row, 2] = float(G_COMy)
        Orbit[ith_row, 3] = float(G_COMz)
        Orbit[ith_row, 4] = float(G_COMvx)
        Orbit[ith_row, 5] = float(G_COMvy)
        Orbit[ith_row, 6] = float(G_COMvz)
       
    #print (Orbit)
    # Save Orbit output files for plotting
    np.savetxt(fileout, Orbit, header = 't x y z vx vy vz', comments = '#', fmt = ['%.2f','%.2f','%.2f','%.2f','%.2f','%.2f','%.2f'])  
        
        
    
        
        
        
        
        
    
    
    
```

```python
"""Making plots"""
import matplotlib.pyplot as plt
%matplotlib inline


"""Load files, seperate parameters, """
MW_data = np.loadtxt("OrbitMW.txt", comments = '#')

# Time for all files
Time = MW_data[:,0] 

# Positions and Velocity
MWx = MW_data[:,1]
MWy = MW_data[:,2]
MWz = MW_data[:,3]
MWvx = MW_data[:,4]
MWvy = MW_data[:,5]
MWvz = MW_data[:,6]


# M31 position and velocity
M31_data = np.loadtxt("OrbitM31.txt", comments = '#')
M31x = M31_data[:,1]
M31y = M31_data[:,2]
M31z = M31_data[:,3]
M31vx = M31_data[:,4]
M31vy = M31_data[:,5]
M31vz = M31_data[:,6]


# M33 position and data
M33_data = np.loadtxt("OrbitM33.txt", comments = '#')
M33x = M33_data[:,1]
M33y = M33_data[:,2]
M33z = M33_data[:,3]
M33vx = M33_data[:,4]
M33vy = M33_data[:,5]
M33vz = M33_data[:,6]


"""Seperations"""
"""Positions for plotting"""
R_MW_31 = abs(np.sqrt((MWx - M31x)**2 + (MWy - M31y)**2 + (MWz - M31z)**2))
R_M33_31 = abs(np.sqrt((M33x - M31x)**2 + (M33y - M31y)**2 + (M33z - M31z)**2))
R_MW_33 =abs(np.sqrt((MWx - M33x)**2 + (MWy - M33y)**2 + (MWz - M33z)**2))
"""Velocity for plotting"""
VR_MW_31 = abs(np.sqrt((MWvx - M31vx)**2 + (MWvy - M31vy)**2 + (MWvz - M31vz)**2))
VR_M33_31 = abs(np.sqrt((M33vx - M31vx)**2 + (M33vy - M31vy)**2 + (M33vz - M31vz)**2))
VR_MW_33 =abs(np.sqrt((MWvx - M33vx)**2 + (MWvy - M33vy)**2 + (MWvz - M33vz)**2))


"""Make Subplot stacked"""

f = plt.figure(figsize = (20,10))

plt.subplot(2,1,1)
plt.subplots_adjust(hspace = 0.35)
plt.title("Galaxy Merger COM Position Seperation", fontsize = 15)
plt.xlabel("Time [Gyr]", fontsize = 15)
plt.ylabel("|$\Delta R_{sep}|$ [Kpc]", fontsize = 15)
#plt.yscale('log')
plt.axis([0, 12, 0, 825])
#plt.axis([6, 12, 0, 50])
#plt.axis([8, 12, 0, 100])
plt.plot(Time, R_MW_31, 'k-', label = "MW/M31 Seperation")
plt.plot(Time, R_M33_31, 'r--', label = "M33/M31 Seperation")
#plt.plot(Time, R_MW_33, 'b.-', label = "MW/M33 Seperation")
plt.legend()
plt.grid()

plt.subplot(2,1,2)
plt.title("Galaxy Merger COM Velocity Seperation", fontsize = 15)
plt.xlabel("Time [Gyr]", fontsize = 15)
plt.ylabel("|$\Delta V|$ [Km/s]", fontsize = 15)
#plt.yscale('log')
plt.axis([0, 12, 0, 600])
#plt.axis([8, 12, 0, 100])
plt.plot(Time, VR_MW_31, 'k-', label = "MW/M31 Seperation")
plt.plot(Time, VR_M33_31, 'r--', label = "M33/M31 Seperation")
#plt.plot(Time, VR_MW_33, 'b.-', label = "MW/M33 Seperation")
plt.legend()
plt.grid()
```
